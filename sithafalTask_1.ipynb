```python
import pdfplumber

# Function to extract text along with page numbers
def extract_text_with_page_numbers(pdf_path):
    extracted_data = []  # List to store tuples of (page_number, text)

    with pdfplumber.open(pdf_path) as pdf:
        for page_number, page in enumerate(pdf.pages, start=1):
            text = page.extract_text()  # Extract text from the page
            if text:
                extracted_data.append((page_number, text))  # Store page number and text as a tuple

    return extracted_data

# Example usage
pdf_path = "Tables, Charts, and Graphs with Examples from History, Economics, Education, Psychology, Urban Affairs and Everyday Life - 2017-2018.pdf"  # Local Colab file path
extracted_data = extract_text_with_page_numbers(pdf_path)

# Print the content of all pages with page numbers
for page_number, text in extracted_data:
    print(f"Page {page_number}:\n{text[:500]}...")  # Print first 500 characters of each page's content
```

```python
def chunk_text_with_page_numbers(extracted_data, max_len=512):
    chunked_data = []  # List to store tuples of (page_number, chunk_text)

    for page_number, page_text in extracted_data:
        # Split the page text into chunks of specified max length
        chunks = [page_text[i:i+max_len] for i in range(0, len(page_text), max_len)]

        # Store each chunk along with its corresponding page number
        for chunk in chunks:
            chunked_data.append((page_number, chunk))

    return chunked_data

# Example usage
chunked_data = chunk_text_with_page_numbers(extracted_data)
# Verify the number of chunks in chunked_data
print(f"Total chunks in chunked_data: {len(chunked_data)}")

# Check the first few chunks
for i, (page_number, chunk) in enumerate(chunked_data[:5]):  # Adjust as needed
    print(f"Chunk {i + 1} | Page {page_number}: {chunk[:100]}...")  # Print the first 100 characters

# Print the first few chunks with their page numbers
#for page_number, chunk in chunked_data[:5]:  # Print first 5 chunks as an example
 #   print(f"Page {page_number}: {chunk[:500]}...")  # Print first 500 characters of each chunk

```

```python
from sentence_transformers import SentenceTransformer

# Load pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')
pdf_path = "Tables, Charts, and Graphs with Examples from History, Economics, Education, Psychology, Urban Affairs and Everyday Life - 2017-2018.pdf"
extracted_data = extract_text_with_page_numbers(pdf_path)
chunked_data = chunk_text_with_page_numbers(extracted_data)

def generate_embeddings_with_page_numbers(chunked_data):
    embeddings_with_page_numbers = []

    for page_number, chunk_text in chunked_data:
        emb = model.encode([chunk_text])[0]  # Generate embedding for the chunk
        embeddings_with_page_numbers.append((page_number, emb))  # Store embedding with page number

    return embeddings_with_page_numbers

# Generate embeddings
embeddings_with_page_numbers = generate_embeddings_with_page_numbers(chunked_data)

# Check embeddings
for i, (page_number, emb) in enumerate(embeddings_with_page_numbers[:5]):  # Print the first 5 embeddings
    print(f"Chunk {i + 1} | Page {page_number}: Embedding shape {emb.shape}")

```

```python
import faiss
import numpy as np

def store_embeddings_with_page_numbers(embeddings_with_page_numbers, index_path='faiss_index_with_page_numbers'):
    # Create a FAISS index using L2 distance
    index = faiss.IndexFlatL2(embeddings_with_page_numbers[0][1].shape[0])
    page_numbers = [page_number for page_number, _ in embeddings_with_page_numbers]
    np.save("page_numbers.npy", page_numbers, allow_pickle=True)  # Use the embedding size for L2 distance
    for _, emb in embeddings_with_page_numbers:
        index.add(np.array([emb]))  # Add each embedding to the index

    faiss.write_index(index, index_path)  # Save the FAISS index
    print(f"Number of embeddings in FAISS index: {index.ntotal}")
# Example usage
store_embeddings_with_page_numbers(embeddings_with_page_numbers)
print("FAISS index has been saved successfully.")
```

```python
# Function to generate query embedding
def query_embedding(query):
    return model.encode([query])

# Function to search the FAISS index with page numbers
def search_index_with_page_numbers(query_embedding, k=3, index_path='faiss_index_with_page_numbers'):
    # Load the FAISS index
    index = faiss.read_index(index_path)

    # Perform the search
    distances, indices = index.search(query_embedding, k)

    return distances, indices

# Example query: asking about types of visualizations
query = "What are the different types of data visualizations?"
query_emb = query_embedding(query)

# Perform the search
distances, indices = search_index_with_page_numbers(query_emb)

# Load the saved page numbers
page_numbers = np.load("page_numbers.npy", allow_pickle=True)

# Print the results
for i in indices[0]:
  if i < len(chunked_data):
        print(f"Page {page_numbers[i]}: {chunked_data[i][1][:100]}...")  # Print first 500 characters of the top chunk

```
